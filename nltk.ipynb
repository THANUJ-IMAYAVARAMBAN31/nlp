```python
import nltk

corpus = """machine learning is a branch of artificial intelligence.
python is a popular programming language.
artificial intelligence helps machines learn."""
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download( 'wordnet')
```

```python
# Tokenize
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize

corpus = corpus.lower() # making everything into lower case
tokens = sent_tokenize(corpus)
tokens
```

```python
# remove stop_words and lemmatize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

for i in range(len(tokens)):
    words = word_tokenize(tokens[i])
    words = [lemmatizer.lemmatize(word , pos ='v') for word in words if word not in stopwords.words("english")]
    tokens[i] = ' '.join(words)

statement = str(tokens)
print(type(statement))
statement
```

```python
# td-idf vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(tokens)
```

```python
# get user input and with cosine similarity retrive data from corpus
user = input("[USER] : ").lower()

# apply same preprocessing
user = lemmatizer.lemmatize(user , pos = 'v')
y = vectorizer.transform([user])

similarity = cosine_similarity(y , X)

index = similarity.argmax()
score= similarity.max()
result = tokens[index]

if score < 0.2:
  print("[BOT] : Sorry , I don't understand")
else :
  print("[BOT] : ", result)
```
